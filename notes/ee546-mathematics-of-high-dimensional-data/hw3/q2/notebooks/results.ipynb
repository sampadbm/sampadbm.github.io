```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib
matplotlib.rc('axes.formatter', useoffset=False)
```

```python
dataset="r20"
```

```python
r2 = pd.read_csv(f"../results/{dataset}U.csv")

plt.subplot(211)
sns.scatterplot(data= r2[r2.eU > 0.1], x="cU", y="eU", color='black', label="> 0.1")
sns.scatterplot(data= r2[r2.eU > 0.05][r2.eU <= 0.1], x="cU", y="eU", color='red', label="0.05-0.1")
sns.scatterplot(data= r2[r2.eU <= 0.05][r2.eU > 0.01], x="cU", y="eU", color="green", label="0.05-0.01")
sns.scatterplot(data= r2[r2.eU <= 0.01], x="cU", y="eU", color="blue", label="< 0.01")
plt.xlabel("cU")
plt.ylabel("eU")
plt.xlim(0,15000)
# plt.savefig(f"{dataset}U.png")


# plt.clf()
plt.subplot(212)
sns.scatterplot(data= r2[r2.eU > 0.1], x="cU", y="eU", color='black', label="> 0.1")
sns.scatterplot(data= r2[r2.eU >= 0.05][r2.eU < 0.1], x="cU", y="eU", color='red', label="> 0.05")
sns.scatterplot(data= r2[r2.eU <= 0.05][r2.eU > 0.01], x="cU", y="eU", color="green", label="0.05 - 0.01")
sns.scatterplot(data= r2[r2.eU <= 0.01], x="cU", y="eU", color="blue", label="< 0.01")
plt.xlabel("cU")
plt.ylabel("eU")
plt.xlim(0,2000)
# plt.ylim(0,0.1)
plt.savefig(f"plots/{dataset}.png")
```

```python
r2 = pd.read_csv(f"../results/{dataset}V.csv")
sns.scatterplot(data= r2[r2.eV > 0.1], x="cV", y="eV", color='black', label="> 0.1")
sns.scatterplot(data= r2[r2.eV >= 0.05][r2.eV < 0.1], x="cV", y="eV", color='red', label="> 0.05")
sns.scatterplot(data= r2[r2.eV <= 0.05][r2.eV > 0.01], x="cV", y="eV", color="green", label="0.05 - 0.01")
sns.scatterplot(data= r2[r2.eV <= 0.01], x="cV", y="eV", color="blue", label="< 0.01")
plt.xlabel("cV")
plt.ylabel("eV")
plt.xlim(0,1000)
plt.savefig(f"plots/{dataset}V.png")
```
